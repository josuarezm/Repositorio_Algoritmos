```python
from queue import PriorityQueue
class Graph:
  def __init__(self, num_of_vertices): #O(n^2)
    self.v = num_of_vertices #O(1)
    self.edges = [[-1 for i in range(num_of_vertices)] for j in range(num_of_vertices)] #O(n^2)
    self.visited = [] #O(1)
  def add_edge(self, u, v, weight): #O(1)
    self.edges[u][v] = weight #O(1)
    self.edges[v][u] = weight #O(1)

g = Graph(22)
g.add_edge(1, 2, 3421)
g.add_edge(1, 3, 5668)
g.add_edge(4, 5, 6012)
g.add_edge(6, 7, 5937)
g.add_edge(3, 8, 6825)
g.add_edge(8, 9, 6062)
g.add_edge(9, 10, 5073)
g.add_edge(11, 12, 3555)
g.add_edge(13, 14, 7291)
g.add_edge(15, 16, 2332)
g.add_edge(17, 18, 3142)
g.add_edge(17, 16, 4342)
g.add_edge(17, 19, 6091)
g.add_edge(5, 19, 5377)
g.add_edge(20, 21, 4541)

# Sevillana: 1
# P. Tunal: 2
# Mandalay: 3
# Venecia: 4
# Marsella: 5
# P. Sur: 6
# P. Américas: 7
# Av. Rojas: 8
# Minuto de Dios: 9
# Suba - Av. Boyacá: 10
# Niza: 11
# Prado: 12
# P. Norte: 13
# 21 Ángeles: 14
# Suba - Cl. 95: 15
# Calle 106: 16
# Avenida 68: 17
# Suba - Cl. 100: 18
# CAN: 19
# Molinos: 20
# P. 20 de Julio: 21

def dijkstra(graph, start_vertex): #O(n^2)
    D = {v:float('inf') for v in range(graph.v)} #O(n)
    D[start_vertex] = 0 #O(1)

    pq = PriorityQueue() #O(1)
    pq.put((0, start_vertex)) #O(1)

    while not pq.empty(): #O(n)
        (dist, current_vertex) = pq.get() #O(1)
        graph.visited.append(current_vertex) #O(1)

        for neighbor in range(graph.v): #O(n^2)
            if graph.edges[current_vertex][neighbor] != -1: #O(n)
                distance = graph.edges[current_vertex][neighbor] #O(1)
                if neighbor not in graph.visited: #O(1)
                    old_cost = D[neighbor] #O(1)
                    new_cost = D[current_vertex] + distance #O(1)
                    if new_cost < old_cost: #O(1)
                        pq.put((new_cost, neighbor)) #O(n)
                        D[neighbor] = new_cost #O(1)
    return D
```

```python
D = dijkstra(g, 1) #O(1)
print(D) #O(1)
```
